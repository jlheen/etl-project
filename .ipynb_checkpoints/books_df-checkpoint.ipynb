```python
import pandas as pd
```

```python
file = 'books.csv'
bookpd_df = pd.read_csv(file)
```

```python
bookpd_df.head()
```

```python
clean_books_df = bookpd_df[['title', 'authors', 'average_rating', 'isbn', '# num_pages']]
clean_books_df.head()
```

```python
isbn_list = clean_books_df['isbn'].tolist()
```

```python
print(isbn_list)
```

```python
import requests
import pandas as pd
import json
from pprint import pprint
```

```python
url = 'https://openlibrary.org/api/books?'
```

```python
book_dict = {}
# need to insert null values where there are no books found
lib_url = []

for isbn in isbn_list:

    try: 
        query = f'{url}bibkeys=ISBN:{isbn}&format=json'
        response = requests.get(query).json()
        lib_url.append(response['ISBN:{}'.format(isbn)]['info_url'])
        
        
    except: 
        print('no book found')
```

```python
# rewrote the code so that dictionary still captured the "null" values
# allows for both arrays (isbn list + book_urls) to be same length, 
# which is necessary for creating a dataframe

books_urls = []

for isbn in isbn_list:

    try: 
        query = f'{url}bibkeys=ISBN:{isbn}&format=json'
        response = requests.get(query).json()
        books_urls.append(response['ISBN:{}'.format(isbn)]['info_url'])
        
        
    except: 
        books_urls.append('no response found')
```

```python
# create a dictionary to hold all responses
books_dict = {
    'isbn': isbn_list,
    'library_url': books_urls
}
```

```python
books_dict
```

```python
books_df = pd.DataFrame.from_dict(books_dict)
books_df
```
